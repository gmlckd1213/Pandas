# Matplotlib

파이썬에서 데이터를 그래프나 차트로 시각화할 수 있는 라이브러리

````python

import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]
fig, ax = plt.subplots() #객체기반스타일로 그린다
# fig -> 그림 개수
# ax -> 그래프개수
plt.plot(x,y)
plt.title("First Plot")
plt.xlabel("x")
plt.ylabel("y")
````

# Line plot

````python
fig, ax = plt.subplots()
x = np.arange(15)
y = x ** 2
ax.plot(
    x, y,
    linestyle=":",
    marker="*",
    color="#524FA1")
````

### Line style

````python
x = np.arange(10)
fig, ax = plt.subplots()
ax.plot(x, x, linestyle="-")
#solid
ax.plot(x, x+2, linestyle="--")
#dashed
ax.plot(x, x+4, linestyle="-.")
#dashdot
ax.plot(x, x+6, linestyle=":")
#dotted
````

### LIne Color

````python
x = np.arange(10)
fig, ax = plt.subplots()
ax.plot(x, x, color="r")
ax.plot(x, x+2, color="green")
ax.plot(x, x+4, color="0.8")
ax.plot(x, x+6, color="#524FA1")
````

### Line Marker

````python
x = np.arange(10)
fig, ax = plt.subplots()
ax.plot(x, x, marker=".")
ax.plot(x, x+2, marker="o")
ax.plot(x, x+4, marker="v")
ax.plot(x, x+6, marker="s")
ax.plot(x, x+8, marker="*")
````

### 축 경계 조정하기

````python
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x))
ax.set_xlim(-2,12)
ax.set_ylim(-1.5,1.5)
````

### 범례

legend

````python
fig, ax = plt.subplots()
ax.plot(x, x, label='y=x')
ax.plot(x, x**2, label='y=x^2')
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend( 
    loc='upper right',# 위치지정
    shadow=True,
    fancybox=True, #모서리 둥글게 
    borderpad=2 #흰색크기
)
````

````python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


#이미 입력되어 있는 코드의 다양한 속성값들을 변경해 봅시다.
x = np.arange(10)
fig, ax = plt.subplots()
ax.plot(
    x, x, label='y=x',
    marker='o',
    color='blue',
    linestyle=':'
)
ax.plot(
    x, x**2, label='y=x^2',
    marker='^',
    color='red',
    linestyle='--'
)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend(
    loc='upper left',
    shadow=True,
    fancybox=True,
    borderpad=2
)



````
